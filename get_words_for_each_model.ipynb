```python
import numpy as np
import pandas as pd

samples = np.genfromtxt('data/samples.csv', delimiter=",").astype(bool)
holdouts = np.genfromtxt('data/holdouts.csv', delimiter=",").astype(bool)
tests = np.genfromtxt('data/tests.csv', delimiter=",").astype(bool)
```

```python
words = pd.read_csv('data/kidwords/kidwords.csv', names=['word'])

words['word_id'] = words.index
subset_data_for_lmem = pd.read_csv('outputs/brute_force_2_with_frequency/subset_data_for_lmem.csv')
model_ids = set(subset_data_for_lmem['model_id'])
```

```python
sample = 82
samples[:, sample].shape
```

```python
len(samples[sample])
```

```python
output_dict = {}

for model_id in model_ids:
    training_set = words[samples[:,sample]]
    training_set['condition'] = "train"

    test_set = words[tests[:,sample]]
    test_set['condition'] = "test"

    holdout_set = words[holdouts[:,sample]]
    holdout_set['condition'] = "holdout"

    df = pd.concat([training_set, test_set, holdout_set], ignore_index=True)
    df['model_id'] = model_id

    output_dict[model_id] = df
```

```python
pd.concat(output_dict.values(), ignore_index=True).to_csv('outputs/brute_force_2_with_frequency/subset_data_for_lmem_items.csv', index=False)
```
