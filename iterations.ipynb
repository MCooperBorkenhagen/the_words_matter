This script takes in the `iterations.csv` file generated by DMK and produces `samples.csv` and `holdouts.csv`. The file `samples.csv` is the version that can be utilized directly by the modeling routine a numpy array and `holdouts.csv` specifies those words (implicit in `samples.csv`) that are heldout during every iteration in the modeling routine. The holdouts can be determined directly by computing the words in `samples.csv` that never appear in the sample (i.e., those rows in that dataframe in which the value is never zero).

```python
import numpy as np
import pandas as pd
```

Sort the iterations data based on the `id` column so that we can sample by index. We will also grab the `holdout` column so that we have that as a separate object (removing it from the `probs` object). The samples will be saved as an array and the holdouts (though redundant) will be saved as a dataframe.

```python
probs = pd.read_csv('data/iterations.csv')
holdout = probs[['id', 'holdout']]
probs = probs.drop('holdout', axis=1)
probs = probs.sort_values(by="id", ascending=True)

# the version
samples = probs[[col for col in probs.columns if 'iter' in col]].to_numpy()
```

Just make sure...

```python
assert list(samples[:, 0]) == probs.iter1.tolist()
```

```python
holdout = holdout.sort_values(by="id", ascending=True)
```

```python
holdout.to_csv('data/holdouts.csv', index=False)
```

```python
np.savetxt('data/samples.csv', samples, delimiter=",")
```
